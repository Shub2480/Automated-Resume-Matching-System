```python
j = 0
stop_words = set(stopwords.words('english'))
for i in os.listdir("New folder"):
    print(i)
    try:
        reader = PdfReader(os.path.join("New folder",i))
    except:
        pass
    
    text=''
    for k in range(len(reader.pages)):
        text += reader.pages[k].extract_text()
    text=text.lower()
    #text = nlp(text)
    tokens = nltk.sent_tokenize(text)
    corpus=''
    for t in tokens:
        
        if 'skill' in t:
            
            corpus += t
    
    corpus=re.sub(r'[^a-zA-Z]', ' ', corpus)
    words=nltk.word_tokenize(corpus)
    

    filtered_words = [word for word in words if word.lower() not in stop_words]
    filtered_words =filtered_words+[" ".join(i) for i in  list(ngrams(filtered_words, 2))]

    result_words=[f for f in filtered_words if process.extract(f,skills,limit=1)[0][1] > 95 ]      
    print(result_words)   
    df=pd.concat([df,pd.DataFrame({'path':i,'skills':[result_words]})],axis=0)   

    print(j)
    j+=1
    
    
```

```python
def remove_duplicate(skills):
    skills=skills.replace("'","").removeprefix('[').removesuffix(']').replace(" ","").split(',')

    skills=list(set(skills))

    skills.sort()

    return ", ".join(skills)
```

```python
df['new_skills']=df.apply( lambda x : remove_duplicate(x[1]),axis=1)
```

```python
df['new_skills']=df['new_skills'].replace('',np.nan)
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.drop('skills',axis=1,inplace=True)
```

```python
df.rename({'new_skills':'skills'},axis=1,inplace=True)
```

```python
from sentence_transformers import SentenceTransformer
```

```python
sbert_model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
file=open('sbert.pkl','wb')
joblib.dump(sbert_model,file)
```

```python
def skill_embeder(skill):
    return sbert_model.encode(skill)
```

```python
df['skill_embeds']=df.apply(lambda x: skill_embeder(x[1]),axis=1)
```

```python
file=open('resume_embeds.pkl','wb')
pickle.dump(df,file)
```
